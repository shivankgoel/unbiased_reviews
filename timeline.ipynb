```python
from helper import *
from datasetcreator import *
from dependencies import *
from loaddata import review_data,business_data, rest_data, user_data, elite_udata, non_elite_udata
```

```python
review_data.sort_values('date', inplace = True)
```

```python
bid_2_stars = review_data.groupby('business_id')['stars'].apply(list).to_dict()
bid_2_dates = review_data.groupby('business_id')['date'].apply(list).to_dict()
bid_2_stars_e = review_data.loc[review_data['iselite'] == True].groupby('business_id')['stars'].apply(list).to_dict()
```

```python
allbids = list(bid_2_stars.keys())
bids = []
for x in allbids:
    if(x in bid_2_stars_e and len(bid_2_stars_e[x]) >= 40 and len(bid_2_stars[x]) >= 120):
        bids.append(x)
print("total bids ", len(bids))
bid = bids[2]
```

```python
%matplotlib qt
def main_function(bids):
    l1 = 120
    y1_s = []
    y2_s = []
    x = [i for i in range(1,l1+1)]

    for bid in bids:
        l = len(bid_2_stars[bid])
        y1 = np.array([x/y for x,y in zip(np.cumsum(bid_2_stars[bid][:l1]),range(1,l1+1))])
        y2 = np.array([np.mean(bid_2_stars_e[bid]) for i in range(1,l1+1)])
        y1_s.append(y1)
        y2_s.append(y2)
        
    y1_s = np.array(y1_s)
    y2_s = np.array(y2_s)
    y1 = np.mean(y1_s,axis=0)
    y2 = np.mean(y2_s,axis=0)
    plt.plot(x,y1,'g-',label='avg ratings over time')
    plt.plot(x,y2,'r--',label = 'elite average rating')
    plt.xticks([r*10 for r in range(1,13)])
    plt.yticks([3.0 + x * 0.2 for x in range(2,8)])
    plt.legend(loc='best')
    return x, y1_s, y2_s

x, y1_s, y2_s =main_function(bids)
```
